## Open CV Evaluation

```python
import cv2
import numpy as np
import urllib
import urllib.request as ur
```

```python
url="https://thumbor.forbes.com/thumbor/960x0/https%3A%2F%2Fspecials-images.forbesimg.com%2Fdam%2Fimageserve%2F1068867780%2F960x0.jpg%3Ffit%3Dscale"
```

```python
image_file="test_image.png"
```

```python
urllib.request.urlretrieve(url, image_file)
```

```python
resp = ur.urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype="uint8")
```

```python
cascPath = "haarcascade_frontalface_default.xml"
faceCascade = cv2.CascadeClassifier(cascPath)
image = cv2.imread("waxobe_2021.jpeg")
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
faces = faceCascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5,
    minSize=(30, 30),
    #flags = cv2.cv.CV_HAAR_SCALE_IMAGE
)
```

```python
print ("Found {0} faces!".format(len(faces)))

# Draw a rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (0, 255, 0), 2)
```

```python
cv2.imshow("Faces found", image)
cv2.waitKey(0)
```

```python
im = cv2.imdecode(image, cv2.IMREAD_COLOR)
blur = cv2.GaussianBlur(im, (5,5),0)
cv2.imshow('image',blur)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
